# Line Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
spotify_data = pd.read_csv("spotify.csv", index_col ="Date", parse_dates = True)
```

```python
spotify_data
```

```python
spotify_data.head()
```

```python
spotify_data.tail()
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(data = spotify_data)
plt.title("Trend in Spotify Songs")
plt.show()
```

```python
b = pd.read_csv("spotify.csv", index_col ="Date", parse_dates = True)
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(data=b["Despacito"], label="Despacito", color="Red")
sns.lineplot(data=b["HUMBLE."], label="HUMBLE.", color="Black")
plt.ylabel("Despacito and Humble")
```

# Bar Plots

```python
flight_data = pd.read_csv("flight_delays.csv", index_col = "Month")
```

```python
flight_data
```

```python
flight_data.head()
```

```python
plt.title("Average Arrival Delay for AA")
sns.barplot(x=flight_data.index, y=flight_data.AA, color = "Blue")
plt.ylabel('Arrival Delay')
```

```python
plt.title("Average Arrival Delay for AS")
sns.barplot(x=flight_data.index, y=flight_data.AS, palette = "Accent")
plt.ylabel('Arrival Delay')
```

```python
dataset = sns.load_dataset('tips')
```

```python
dataset
```

```python
sns.barplot(x='sex', y='total_bill', data=dataset)
```

```python
sns.barplot(x='sex', y='total_bill', hue = 'day', data=dataset)
```

# Heatmap

```python
plt.figure(figsize=(12,8))
sns.heatmap(data=flight_data, annot = True, cmap ='Greens_r')
```
